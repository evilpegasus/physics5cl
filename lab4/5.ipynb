```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

At an intuitive level, photocurrent should be positively correlated with intensity of incident light, assuming that the threshold voltage is reached. This is because at higher intensities of light, there are simply more photons hitting the electrons on the metal, thus amplifying the photoelectric effect.



```python
# using red and blue LEDs
red = {
    "voltage": [17,19,22,24,26,28,30,33,35,37,39],
    "current": [27,180,413,639,832,1000,1157,1285,1409,1518,1620]
}

blue = {
    "voltage": [16,18,20,23,25,28,29,31,34,36,38,40],
    "current": [89,1043,2767,4174,5216,6394,6729,7322,7832,8294,8721,9096]
}
df_red = pd.DataFrame(red)
df_blue = pd.DataFrame(blue)
```

```python
plt.scatter(df_red["voltage"], df_red["current"])
plt.xlabel("Intensity (% of maximum)")
plt.ylabel("Current (mA)")
plt.title("Photocurrent vs Intenstiy for Red LED");
```

```python
lr = LinearRegression()
lr.fit(df_red[["voltage"]], df_red[["current"]])

plt.scatter(df_red["voltage"], df_red["current"], label="Data Points")
plt.plot(df_red["voltage"], lr.predict(df_red[["voltage"]]), label="Linear Fit", c="r")
plt.xlabel("Intensity (% of maximum)")
plt.ylabel("Current (mA)")
plt.title("Photocurrent vs Intenstiy for Red LED")
plt.legend();
```

```python
plt.scatter(df_blue["voltage"], df_blue["current"])
plt.xlabel("Intensity (% of maximum)")
plt.ylabel("Current (mA)")
plt.title("Photocurrent vs Intenstiy for Blue LED");
```

```python
lr = LinearRegression()
lr.fit(df_blue[["voltage"]], df_blue[["current"]])

plt.scatter(df_blue["voltage"], df_blue["current"], label="Data Points")
plt.plot(df_blue["voltage"], lr.predict(df_blue[["voltage"]]), label="Linear Fit", c="r")
plt.xlabel("Intensity (% of maximum)")
plt.ylabel("Current (mA)")
plt.title("Photocurrent vs Intenstiy for Blue LED")
plt.legend();
```
